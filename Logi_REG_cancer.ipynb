```python
df = pd.read_csv('breast-cancer_csv.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['node-caps'] = le.fit_transform(df['node-caps'])
df['breast'] = le.fit_transform(df['breast'])
df['irradiat'] = le.fit_transform(df['irradiat'])
df['breast-quad'] = le.fit_transform(df['breast-quad'])
df['age'] = le.fit_transform(df['age'])
df['Class'] = le.fit_transform(df['Class'])
df['tumor-size'] = le.fit_transform(df['tumor-size'])
```

```python
X = df.drop(columns = ['menopause','irradiat','inv-nodes'])
Y = df['irradiat']
```

```python
df.head(10)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
```

```python
LR.fit(X_train, y_train)
```

```python
y_predicted = LR.predict(X_test)
```

```python
y_predicted
```

```python
y_prob=LR.predict_proba(X_test)
y_prob[0:10]
```

```python
LR.score(X_test,y_test)
```
